###4 - Extraia e salve os frames 3,5,7,9,11,13,15,....,97 dos vÃ­deos presentes no diretÃ³rio Videos. Os frames deverÃ£o ser salvos no diretorio Frame_exe_4

```python
# Directory
directory = "Faces_exe_4/"
  
# Parent Directory path
parent_dir = "/content/drive/MyDrive/Atividade Avaliativa/"
  
# Path
path_ex4 = os.path.join(parent_dir, directory)
  
# Create the directory

if not os.path.exists(path_ex4):
  os.mkdir(path_ex4)
```

```python
videos = os.listdir('/content/drive/MyDrive/Atividade Avaliativa/Videos')
print(videos)
```

```python
# Bibliotecas
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

##Arquivo do video
for x in range(len(videos)):
  video_path = '/content/drive/MyDrive/Atividade Avaliativa/Videos/' + videos[x]

  #######Leitura do Video
  cap = cv2.VideoCapture(video_path)

  ####Recupera a taxa de frames por segundo
  fps = cap.get(cv2.CAP_PROP_FPS)

  cont = 1 #Contador para controlar a quantidade de frames contido em 1 segundo.
  cont_frame = 3
  cont_max_frame = 11# Quantidade maxima de frames a ser capturados

  ####Modifique o caminho do arquivo de acordo com o seu#######
  caminho_saida = path_ex4


  #######Verifica se o video foi devidamente carregado######
  if cap.isOpened():

      frame = None
      while True:
          try:
              #Captura um frame do video
              is_success, frame = cap.read()
              
              #Transforma a imagem em escala de cinza             
              image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

              #######Recupera uma imagem por segundo######
              if cont > fps:
                if cont_frame <= cont_max_frame:

                    ####DefiniÃ§Ã£o do nome do arquivo e o caminho a ser salvo na rede
                    file_path = caminho_saida + videos[x] + 'Frame_'+ str(cont_frame) + '.jpg'

                    ######Salvando o frame em uma imagem com formato jpg
                    cv2.imwrite(file_path,image)

                    ####Contador que controla a quantidade de frames salvos na rede
                    cont_frame = cont_frame + 2
                cont = 1     

              cont = cont + 1  
          except cv2.error:
              pass

          if not is_success:
              break        

      if frame:
        cv2_imshow(frame)


  cap.release()

  cv2.destroyAllWindows()
```
